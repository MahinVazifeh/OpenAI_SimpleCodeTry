```python
import os
import openai
```

```python
from openai import OpenAI
client = OpenAI(api_key="sk-OpenAI_Key")
```

```python
def create_shopping_list(recepie):
    prompt = f"Give me the main symtoms of MS progression in {(recepie)} type"

    return prompt 
```

```python
# Function list
recepie = create_shopping_list("RRMS")
print(recepie)
```

```python
# Reaching out to Open AI for finding the answer
response = client.chat.completions.create(
    model = "gpt-3.5-turbo",
    messages= [
        {
            "role":"system",
            "content": recepie
        },
    ],
    temperature=0.7,
    top_p=1,
)
print(response.choices[0].message.content)
```
